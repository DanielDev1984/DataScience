### saving to / loading from csv

```python
animals = pd.DataFrame({'Cows' : [12,20], 'Goats' : [22,19]}, index=['Year 1', 'Year 2'])
animals
```

```python
filename = 'animals.csv'
animals.to_csv(filename)
```

```python
# test whether saving was succesful
animals_fromcsv = pd.read_csv(filename, index_col=0)
animals_fromcsv
```
